```python
#loading the dataset
# make sure that the dataset and the ipynb file are in the same damn folder

data = pd.read_csv ('train-2.csv') #package panda, function "read csv", filename
```

```python
# Encode

data ['Sex'] = data['Sex'].astype('category')
data ['Sex'] = data['Sex'].cat.codes
```

```python
data.head()
```

```python
#imputation
# process of reaplacing missimng information with a value
# the value can be an average, median, mode, or you can perfomr regression to predict the missing values

#missing values: deleted (rows, columns), Imputed
#inputation beyond a point reduces the deviation in the data

from sklearn.impute import SimpleImputer
imputer = SimpleImputer (missing_values = np.nan, strategy = "mean") 
# Numpy reads the "nan" as msising data, strategy = mean
imputer = imputer.fit(data[['Age']])
data[['Age']] = imputer.transform(data[['Age']])
data.info()
```

```python
# Plotting a histogram to analyze age

ply.hist(data['Age'], bins=10)

ply.show()
```
